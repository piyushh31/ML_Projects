```python
import numpy as np

first = []
first = np.random.uniform(2,4,[100,1000])
third = []
third = np.random.uniform(0,2,[100,1000])
```

```python
third.shape
```

```python
first_cd = np.mean(first,axis = 0)
third_cd = np.mean(third,axis = 0)
```

```python
point = np.random.uniform(0,4,[1000])
dist1 = np.linalg.norm(point-first_cd) 
dist2 = np.linalg.norm(point-third_cd) 
dist1,dist2
```

```python
if dist1<=dist2: 
    print(f"{point} belongs to cluster 1")
else: print(f"{point} belongs to cluster 2")
```
