```python
import numpy as np
import pandas as pd
df=pd.read_csv("Iris-dataset.csv")
```

```python
dict_map = {}
index = 1
for i in set(df['y']):
    dict_map[i] = index
    index += 1
dict_map
```

```python
np_df = np.zeros((150,4))
y = []
for i in range(len(df)):
    np_df[i] = df.iloc[i][:4]
    y.append(dict_map[df.iloc[i][-1]])
y_test=np.append(np.append(y[40:50],y[90:100],axis = 0),y[140:150],axis = 0)
```

```python
train_data= np.append(np.append(np_df[:40],np_df[50:90],axis = 0),np_df[100:140],axis = 0)
train_data
```

```python
test_data = np.append(np.append(np_df[40:50],np_df[90:100],axis = 0),np_df[140:150],axis = 0)
test_data
```

```python
mean1 = np.mean(train_data[:40],axis = 0)
mean2 = np.mean(train_data[40:80],axis = 0)
mean3 = np.mean(train_data[80:120],axis = 0)
print(mean1,mean2,mean3)
```

```python
def tester(test):
   
    r = []
    for i in test:
        dist1 = np.linalg.norm(i-mean1)
        dist2 = np.linalg.norm(i-mean2)
        dist3 = np.linalg.norm(i-mean3)
        r.append([dist1,dist2,dist3,min([dist1,dist2,dist3])])
        
    return r 
```

```python
a=np.array(tester(test_data))
t=[]
for i in range(30):
    
    if a[i][0]==a[i][3]:
        t.append(2)
    elif a[i][1]==a[i][3]:
        t.append(3)
    else:
        t.append(1)
a
```

```python
l1=[]
for i in range(30):
    if t[i]==y_test[i]:
        l1.append(1)   #1 for accurate
    else:
        l1.append(0)   #0 for inaccurate
l1=np.array(l1)
l1
```

```python
accuracy=np.sum(l1)/len(a)*100
accuracy
```
