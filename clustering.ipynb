```python
import numpy as np
first=[]
for i in range(100):
    first.append([np.random.uniform(2,4),np.random.uniform(2,4)])
third=[]
for i in range(100):
    third.append([np.random.uniform(0,2),np.random.uniform(0,2)])

import matplotlib.pyplot as plt
plt.scatter([i[0] for i in first],[i[1] for i in first])
plt.scatter([i[0] for i in third],[i[1] for i in third])
```

```python
m1=np.mean(first,axis=0)
m2=np.mean(third,axis=0)
print(m1,m2)
```

```python
import matplotlib.pyplot as plt
plt.scatter([i[0] for i in first],[i[1] for i in first])
plt.scatter([i[0] for i in third],[i[1] for i in third])
plt.scatter([m1[0]],[m1[1]])
plt.scatter([m2[0]],[m2[1]])
```

```python
def test(x,y):
    dist1=np.linalg.norm(m1-[x,y])
    dist2=np.linalg.norm(m2-[x,y])
    
    if dist1<=dist2:
        print("x,y lies in cluster 1")
    else:
        print("x,y lies in cluster 2")
    
    print("dist1 and dist2 from points respectively:")
    return dist1,dist2

test(2,2)
```

```python
test(3,3)
```

```python
test(1,2.5)
```
